Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.

Matplotlib is a widely-used plotting library for the Python programming language. It provides an object-oriented API for embedding plots into applications that use graphical user interfaces and is also popular for generating plots in interactive Python sessions or Jupyter notebooks.

Why it is used:

Data Visualization: It allows users to create a wide variety of static, animated, and interactive plots to visualize data effectively.

Customizability: Matplotlib provides extensive customization options for plot appearance, making it versatile for different types of visualizations.

Integration: It works well with various data analysis libraries, like NumPy and pandas, and is compatible with multiple platforms.

Publication Quality: It can generate high-quality plots suitable for inclusion in publications.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.random.rand(10)
```

```python
y=np.random.rand(10)
```

```python
plt.scatter(x,y)
plt.xlabel('this is my x data')
plt.ylabel('this is my y data')
```

```python
plt.plot(x,y)
plt.xlabel('this is my x data')
plt.ylabel('this is my y data')
plt.title('2D rep')
```

```python
plt.bar(x,y)
```

Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
Note: Also add title, xlabel, and ylabel to the plot.


A scatter plot is a type of data visualization used to represent the relationship between two continuous variables. Each point in a scatter plot corresponds to one observation in the dataset, and the coordinates of the point represent the values of the two variables

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
x
```

```python
y
```

```python
len(x)
```

```python
colours=np.random.rand(50)
plt.scatter(x,y,c=colours)
plt.xlabel('x-axis',)
plt.ylabel('y-axis')
plt.title('scatter plotting of data')
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])


A bar plot (or bar chart) is a type of data visualization that displays data using rectangular bars. Each bar's length represents the value of a particular category, making it easy to compare different categories.

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.bar(company,profit)
```

```python
plt.barh(company,profit)
```
